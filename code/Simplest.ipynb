We first create a REBOUND simulation, then we add two particles and integrate the system for 100 time units.

```python
import rebound
sim = rebound.Simulation()
```

```python
sim.add(m=1.)
sim.add(m=1e-3, a=1., e=0.1)
sim.add(a=1.4, e=0.1)
```

```python
sim.integrate(100)
```

```python
#  输出在100个时间单位的积分后三个粒子的位置参数
for i in range(0, sim.N):
    p = sim.particles[i]
    print(p.x, p.y, p.z)
```

```python
primary = sim.particles[0]

# 输出在100个时间单位的积分后两个粒子的半长轴、轨道偏心率以及真近点角参数
for i in range(1, sim.N):
    p = sim.particles[i]
    o = p.calculate_orbit(primary=primary)
    print(o.a, o.e, o.f)
```

```python
# 输出图像
%matplotlib inline
fig = rebound.OrbitPlot(sim, unitlabel="[AU]")
```
